```python
import pandas as pd
import numpy as np
```

# Add number of restaurants

```python
population = pd.read_csv("population.csv")
fast = pd.read_csv("FastFoodRestaurants-csv.csv")
veggie = pd.read_csv("vegetarian_restaurants-csv.csv", encoding='latin-1')
```

```python
veggie.head()
```

```python
fast.columns.values
```

```python
fast["province_code"].unique()
```

```python
population.head()
```

```python
veggie["veggie_num"]=1
veggie_count = veggie.groupby("province_code").agg("count")["veggie_num"]
veggie_count = veggie_count.to_frame()
veggie_count["NAME"] = veggie_count.index
veggie_count.head()
```

```python
fast["fast_num"]=1
fast_count = fast.groupby("province_code").agg("count")["fast_num"]
fast_count = fast_count.to_frame()
fast_count["NAME"] = fast_count.index
fast_count.head()
```

```python
re_pop = pd.merge(population, veggie_count, left_on = "NAME", right_on = "NAME")
re_pop.head()
```

```python
re_pop = pd.merge(re_pop, fast_count, left_on = "NAME", right_on = "NAME")
re_pop.head()
```

# Add division information

```python
division = pd.read_csv("divisions.csv")
```

```python
division.head()
```

```python
re_pop = pd.merge(re_pop, division[["State","Region"]], left_on="NAME", right_on = "State")
```

```python
del re_pop["State"]
re_pop.to_csv("processed_population.csv", sep=",")
```
